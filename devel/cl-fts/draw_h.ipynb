```python
import json
import copy
import pathlib
import pickle
import glob
import os
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import savemat, loadmat

folder_name = "data_simulation_no_ds"

data_list = sorted(glob.glob(f"{folder_name}/dH*.mat"))
print(data_list)

hamlitonians = np.array([])
for data in data_list:
    hamlitonians = np.concatenate((hamlitonians, loadmat(data, squeeze_me=True)["H_history"]))

# 
size = len(hamlitonians)
re_data = np.real(hamlitonians[round(size*0.2):])
im_data = np.imag(hamlitonians[round(size*0.2):])
# x = np.linspace(0, 10, round(size*0.9))
plt.plot(re_data)
plt.ylabel("Re[H]")
plt.show()

plt.plot(im_data)
plt.ylabel("Im[H]")
plt.show()


plt.hist(re_data, bins=30, edgecolor='black', alpha=0.7) 
plt.show()

plt.hist(im_data, bins=30, edgecolor='black', alpha=0.7) 
plt.show()
```
