```python
import numpy as np
import json
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import scipy.io as sio

data = sio.loadmat("phases/C2D.mat", squeeze_me=True)
# data = json.load(open("phases/C2D.json", "r"))

nx = data['nx']
lx = data['lx']
phi_A = np.array(data['phi_A']).reshape(nx)
phi_B = np.array(data['phi_B']).reshape(nx)

fig, axes = plt.subplots(1, 2, figsize=(8, 4))
fig.suptitle("Concentration" , fontsize=16)
im = axes[0].imshow(phi_A, extent=(0, lx[1], 0, lx[0]), origin='lower', cmap=cm.jet, vmin=0.0, vmax=1.0)
im = axes[1].imshow(phi_B, extent=(0, lx[1], 0, lx[0]), origin='lower', cmap=cm.jet, vmin=0.0, vmax=1.0)
axes[0].set(title='phi_A', xlabel='y', ylabel='x')
axes[1].set(title='phi_B', xlabel='y', ylabel='x')

fig.subplots_adjust(right=1.0)
fig.colorbar(im, ax=axes.ravel().tolist())
fig.show()

# fig.savefig("2d_density.png", dpi=300, bbox_inches='tight', pad_inches=0)
```
